```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# File for data file
file_path = "~/Documents/BedTracks/git_test/Crime_Data_from_2020_to_Present.csv"
```

```python
# Reading .csv file
dataframe = pd.DataFrame(pd.read_csv(file_path))
# Formatting date-time columns
dataframe["Date Rptd"] = pd.to_datetime(dataframe["Date Rptd"], format='%m/%d/%Y %H:%M:%S %p')
dataframe["DATE OCC"] = pd.to_datetime(dataframe["Date Rptd"])
```

```python
# Showing top rows 
pd.set_option("display.max_columns", len(dataframe))
pd.set_option("display.max_rows", 4)
dataframe
```

```python
# Reseting viz options
pd.reset_option("display.max_columns")
pd.reset_option("display.max_rows")
```

```python
# Top locations where crimes took place and total # of crimes
pd.DataFrame(dataframe["Premis Desc"].value_counts()).head(10)
```

```python
# Weapons used, total # of crimes, and percentage from total crimes
weapons = pd.DataFrame(dataframe["Weapon Desc"].value_counts())
weapons = weapons.reset_index().rename(columns={"index":"Weapon","Weapon Desc":"Number of Crimes"})
weapons["Percentage"] = pd.Series(map(lambda x: np.round((x/sum(weapons["Number of Crimes"]))*100,2), 
                                      weapons["Number of Crimes"]))
```
